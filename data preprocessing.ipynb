```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
```

```python
data = pd.read_csv("./data/mud_props_UTF.csv")
```

```python
data
```

```python
data.isna().sum()
```

```python
data.columns
```

```python
feature = data[['Mud Wt. In (ppg)', 'Funnel Viscosity ']]
target = data[['Fann 600 rpm', 'Fann 300 rpm']]
```

```python
feature.shape
```

```python
feature.isnull().sum()
```

```python
numeric_columns = target.select_dtypes(include=[np.number]).columns
print("Numeric columns:", list(numeric_columns))
```

```python
for col in numeric_columns:
    target[col] = target[col].fillna(target[col].mean())
```

```python
import joblib
scaler = MinMaxScaler()
feature_normalized = pd.DataFrame(scaler.fit_transform(feature), columns=feature.columns)
joblib.dump(scaler, 'normalization_feature.pkl')
```

```python
import joblib
scaler = MinMaxScaler()
target_normalized = pd.DataFrame(scaler.fit_transform(target), columns=target.columns)
joblib.dump(scaler, 'normalization_target.pkl')
```
